```python
import pandas as pd
from pathlib import Path
from src.mevpt import *
from src.scaling_helper import *
from src.mplstyle import set as set_mplstyle
set_mplstyle("Fira Sans")

DATA_DOC_PATH = Path("./data/dataset_docs/")
OUTPUT_PATH = Path("./output/")
```

# NBINs = 12

```python
NBINS = 12
```

```python
TMEAN_OE_PATH = Path("./data/OE_tmean_era5/")
STATION_LIST = pd.read_csv(DATA_DOC_PATH / "pt_stations_ge30y_80pct_good.csv")
for var, xlabel in zip(["TMEAN","W500","VIMC"], ["$T~[^{\circ} C]$", "$\omega_{500}~[Pa/s]$", "$VIMC~[mm]$"]):
    plot_scaling(
        var, xlabel, 
        station_list = STATION_LIST,
        oe_dir=TMEAN_OE_PATH, 
        ntimes=1000,
        savepath=OUTPUT_PATH / f"figures/Nbins_{NBINS}_{var}.pdf", 
        n_bins=NBINS)
```

```python
DPT_OE_PATH = Path("./data/OE_dpt_era5/")
STATION_LIST = pd.read_csv(DATA_DOC_PATH / "p_dpt_stations_ge30y_70pct_good.csv")
for var, xlabel in zip(["DPT"], ["$\\text{DPT}~[^{\circ} C]$"]):
    plot_scaling(
        var, xlabel, 
        station_list = STATION_LIST,
        oe_dir=DPT_OE_PATH, 
        ntimes=1000,
        savepath=OUTPUT_PATH / f"figures/Nbins_{NBINS}_{var}.pdf", 
        n_bins=NBINS)
```
